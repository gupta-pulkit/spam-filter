```python
import pandas as pd
import numpy as np
```

```python
f = open("spambase/spambase.data", "r")
```

```python
fl = f.readlines()
print(type(fl))
```

```python
data = []
for a in fl:
    a = a.strip('\n')
    a = a.split(',')
    a = [float(i) for i in a]
    data.append(a)
```

```python
print(len(data))
```

```python
dataset = np.array(data)
```

```python
print(dataset.shape)
```

```python
pd.DataFrame(dataset).to_csv("dataset.csv", index=False)
```
